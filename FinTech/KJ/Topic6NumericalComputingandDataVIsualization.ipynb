### $\bullet$ Simple plot

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_formats = ['svg']
x = [0, 3, 8]
y = [0, 7, 12]

x2 = np.arange(0, 8, 0.01)
y2 = np.sin(x2)

plt.plot(x, y, 'o-.',lw=2, markersize=5, mec='#e00ba4', markerfacecolor='#e00ba4', label='Time&Dist' )
plt.plot(x2, y2, 'r', lw=2, label='Sin(x)')
plt.text(3.2, 7, 'optimal point', fontsize=6)
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Time vs. Distance')
plt.legend()
plt.grid(True)
```

```python
import numpy as np

plt.style.use('default')
x = np.arange(-3, 3, 0.01)
y = (1/(2*np.pi)**0.5)*np.exp(-(x**2)/2)
fig = plt.figure(figsize=(6, 4))
plt.xlabel('x')
plt.title('Standard Normal Distribution')
plt.ylabel(r'$\frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}}$')
plt.plot(x, y)
```

### Barplot

### subplots

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots(2,2)
ax[0,0].plot(t, s)

ax[0,0].set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax[0,0].grid(1)
# plt.subplots_adjust(hspace=0.3)
x2 = np.arange(-np.pi, np.pi, 0.01)
y2 = np.cos(x2)
ax[0,1].plot(x2,y2)

plt.show()
```
