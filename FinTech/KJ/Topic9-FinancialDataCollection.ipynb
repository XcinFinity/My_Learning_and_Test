```python
columns = ['Stkcd',  'Accper', 'F050501B']
condition = "Indcd ='C36' OR Indcd ='C38'"
condition = "Indcd IN ('C36', 'C38') "
data = csmar.query_df(columns, condition, tb, '2020-01-01', '2023-12-31')
data
```
