```python
def PricetoCategory(row):
    price = row['SalePrice']
    if price >= 5e5:
        category = '豪宅'
    elif price >= 2e5:
        category = '普通'
    else:
        category = '低端'
    return category
```

```python
df4['Category'] =  df4.apply(PricetoCategory, axis=1)
df4['category'] =  df4.apply(PricetoCategory, axis=1)
df4.head()
```
