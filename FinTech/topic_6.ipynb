## Data Visualization

```python
# matplotlib - 基础画图
# seaborn - 美观
# plotly - 网站动态图

import matplotlib.pyplot as plt
import numpy as np

%config InlineBackend.figure_format = 'svg'
```

```python
x1 = [1, 3.2, 7.8, 8.4, ]
y1 = [2, 4.3, 7.5, 8.1]
plt.plot(x1, y1, color='blue', marker='o')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x1, y1, 'o', ls='-.', lw='2', c='red', mfc='yellow', label='f(x)')
x2 = np.arange(1, 10, 0.1)
y2 = np.sin(x2)
plt.plot(x2, y2, '-', lw='2', label='sin(x)')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Time vs Distance')
```

```python
x3 = np.arange(-5, 5, 0.01)
y3 = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x3 ** 2)
plt.plot(x3, y3, lw='2', label='normal distribution')

plt.legend()
plt.xlabel('x')
plt.ylabel(r'$ f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} $')
plt.title('Normal Distribution')
plt.savefig('normal_distribution.pdf')
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(10, 5))

plt.plot(x1, y1, 'o', ls='-.', lw='2', c='red', mfc='yellow', label='f(x)')

plt.plot(x2, y2, '-', lw=2, label='sin(x)')

plt.plot(x3, y3, lw='2', label='normal distribution')

plt.legend()
plt.xlabel('x')
plt.ylabel(r'$ f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} $')
```
